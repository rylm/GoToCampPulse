```python
mongo = MongoClient(host='178.62.219.128', port=8888)
db = mongo.main
```

```python
mdata = list(db.user_stats.find())
```

```python
len(mdata)
```

```python
for_graph = []
for record in mdata:
    for key in record:
        if key != '_id':
            for user in record[key]:
                for_graph.append([int(key), user['first_name'] + ' ' + user['last_name'], user['online']])
```

```python
user_set = sorted(list(set([a[1] for a in for_graph])))
```

```python
online = []
online_numbers = []
for user in user_set:
    online.append(list(filter(lambda x: x, [int(a[0]) if a[1] == user and a[2] == 1 else None for a in for_graph])))
    online_numbers.append([a[2] for a in for_graph if a[1] == user])
```

```python
len(online_numbers)
```

```python
online_strings = []
for user in online_numbers:
    online_strings.append(''.join([str(a) for a in user]))
```

```python
%%time
distances = []
for user1 in online_strings:
    for user2 in online_strings:
        if user1 == user2:
            print("I")
            continue
        ldistance = distance(user1, user2)
        distances.append(ldistance)
```

```python
min(distances)
```

```python
len(distances)
arr = np.zeros(shape=(46,46))
index = 0
for i in range(46):
    for j in range(46):
        if i != j:
            arr[i, j] = distances[index]
            index += 1
        else:
            arr[i, j] = 50000
            
print(arr)
```

```python
index = 0
for i in arr:
    closest = np.argmin(i)
    print(user_set[index], " - ", user_set[closest], i[closest])
    index += 1
    
```
